- numpy는 특히 벡터 및 행렬 연산에 있어 편의성을 제공하는 라이브러리

```python
import numpy as np
```

# zeros(), ones(), arange() 함수

```python
# zeros는 0으로 된 array를 만들어주는 것이다. (행과 열을)
np.zeros((3,6))
```

```python
# ones: 1로 된 array를 만든다.
np.ones(10)
```

```python
# arange: 연속된 수로 array를 만든다.
np.arange(15)
```

# 데이터 타입

#### numpy에서 사용 가능한 데이터형 (dtype) 정리

```python
# dtype: 데이터 타입 확인
arr1.dtype
```

```python
arr2.dtype
```

```python
# arr 만들면서 데이터 타입 지정
arr = np.array([1,2,3,4,5], dtype=np.int64)
arr
```

```python
arr.dtype
```

```python
# int형을 float 타입으로 변경 (기존의 타입을 변경 = astype())
float_arr = arr.astype(np.float64)
float_arr.dtype
```

# array 관련 연산

두 array 간에는 더하기, 빼기, 곱하기, 나누기 등의 연산을 수행할 수 있다. '+', '-', '*', '/' 등과 같이 일반 숫자에 대해 사용하는 연산자를 사용. 이 때, 이러한 더하기, 빼기, 곱하기, 나누기 등의 연산은 두 array 상의 동일한 위치의 성분끼리 이루어지게 되므로, 당연히 두 array의 모양이 같아야 계산이 가능하다.

요약: 같은 열에 있는 넘들이랑 연산됨.

```python
arr1 = np.array([[1,2,3], [4,5,6]], dtype=np.float64)
arr2 = np.array([[7,8,9], [10,11,12]], dtype=np.float64)
```

```python
arr1
```

```python
arr2
```

```python
# 2차원에서 같은 위치의 성분끼리 연산이 된다.
arr1 + arr2
```

```python
arr1 * arr2
```

```python
arr1 ** arr2
```

```python
1 / arr1
```
