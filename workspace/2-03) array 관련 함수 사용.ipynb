# 정렬 함수 및 기타 함수

```python
arr = np.random.randn(10)
arr
```

```python
# 10개 데이터 오름차순 정렬
np.sort(arr)
```

```python
# 내림차순 정렬
np.sort(arr)[::-1]
```

```python
# 2차원 array 정렬
arr = np.random.randn(5,3) #5행3열
arr
```

2차원 array를 정의, axis=0을 설정하면 행방향으로 오름차순 정렬, axis=1 설정하면 열방향으로 오름차순 정렬이 수행된다.

```python
# 행방향으로 열의 값을 오름차순 정렬
np.sort(arr, axis=0)
```

```python
# 열방향으로 행의 값을 오름차순 정렬
np.sort(arr, axis=1)
```

```python
# 길이가 아주 긴 array가 있을 때, 상위 5%에 위치하는 값을 찾기 (이런 것도 많이 한다)
1. 데이터 생성 및 내림차순 정렬
2. 5%에 해당하는 위치값을 찾고, 해당하는 값을 가져오면 된다.
```

```python
large_arr = np.random.randn(150)
large_arr
```

```python
# 내림차순 정렬
np.sort(large_arr)[::-1]
```

```python
# 상위 5%에 해당하는 값 찾기
# 0.05*150 # 7.5
# 내림차순된 array에서 0.05*150 을 수행하여 int값으로 변경. 계산된 7번째 자리값이 상위 5%의 값이다.
np.sort(large_arr)[::-1][int(0.05*len(large_arr))]
```
