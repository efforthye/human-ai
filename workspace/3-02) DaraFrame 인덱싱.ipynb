```python
import pandas as pd
import numpy as np
```

```python
# 데이터 프레임을 하나 만들어 둔다.
data = {
    "names": ['aaa','aaa','bbb','ccc','ccc'],
    "year": [2014,2015,2016,2015,2016],
    "points": [1.5, 1.7, 3.6, 2.4, 2.9]
}
df = pd.DataFrame(data, columns=['year', 'names','points','penalty'], index=['one','two','three','four','five'])
df
```

# 행 선택하는 권장 방법: .loc(), .iloc()
- loc: 컬럼 명 기준 인덱싱
- iloc: 자리수 기준 인덱싱


```python
# 
df.loc['two'] # 인덱스 가 two인 행의 전체 값을 가져온다.
```

```python
df.loc['two':'four'] # 문자를 기준으로 범위 인덱싱
```

```python
# points 컬럼의 특정 값만 가져오기 - loc 형태 인덱싱
df.loc['two':'four', 'points'] # 행범위인덱싱, 열값은 points
```

```python
# 전체 행 값 가져오고, 거기에서 year컬럼의 값만 가져오기!!
df.loc[:, 'year']
```

```python
df.loc[:, ['year', 'names']] # year와 names 에 대한 전체 값 가져옴.
```

```python
# 2~3행에 대해 year, names 가져옴.
df.loc['two':'four', ['year','names']]
```

```python
#df.loc['six',:] = [2017, 'ddd', 3.3, 0.6, -1.3]
#df
```

```python
# .iloc는 numpy 방식으로 (행/열 자리값으로)
```

```python
df.iloc[3]
```

```python
# 0,1,2까지 0~3행값(3개), 0,1까지 0~2열값(2개) 가져온다.
df.iloc[0:3, 0:2]
```

```python
# 가끔 0,1,3행 이렇게 가져오고 싶을 때, 행과 열의 범위를 별도 지정해서 인덱싱
df.iloc[[0,1,3], [1,2]]
```

```python
df.iloc[:,1:4]
```

```python
# 값 하나만 가져오기
df.iloc[1,1] #1행1열
```

# 불리언 인덱싱

```python
# year의 2014년보다 큰 값 찾기 (연도에서 2014년보다 큰값)
df.year>2014
```

```python
## add컬럼 만들고 거기에 2014보다 크면 true를 넣음
add = df.year>2014
df['add'] = add
df.loc[:]
```

```python
## 불리언 값에 의해 인덱싱
df.loc[df.year>2014, :] #year가 2014보다 큰거 전체 뽑음
```

```python
## names가 ccc에 해당되는 names와 points정보 뽑아오기
df.loc[df.names=='ccc', ['names', 'points']]
```

```python
## points 값 2보다 크고 3보다 작은 값 인덱싱
df.loc[(df.points>2) & (df.points<3), :]
```

```python
## 포인트값이 3보다 큰것은 penalty 값을 0으로 대임
df.loc[df.points>3, 'penalty']=0
df
```
