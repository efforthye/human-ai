```python
# 문제) 사용자 별 평균 평점 계산하기 - 정답
```

```python
# user에 대해 unique 값만 먼저 다 저장한다.
user_id_arr = np.unique(data[:,0])
user_id_arr
```

```python
# 유저 몇개인지 확인
user_id_arr.shape
```

```python
# 빈 리스트를 생성 -> 여기에 연산해서 나오는 결과 넣을 것
mean_rating_by_user_list = []

for user_id in user_id_arr:
    data_for_user = data[data[:,0]==user_id, :] # 해당 유저의 데이터. 특정 USER_ID 값들의 ARRAY를 저장한 변수 (unique 값 별로 저장될것이다.)
    mean_rating_for_user = data_for_user[:, 2].mean() # 해당 id 에 대한 평균 평점을 계산한다.
    mean_rating_by_user_list.append([user_id, mean_rating_for_user]) # 빈 리스트에 결과를 넣는다.
```

```python
# csv 파일에 결과 저장
```

```python
# 리스트 => array 타입으로 변환
mean_rating_by_user_list_array = np.array(mean_rating_by_user_list, dtype=np.float64) # array 타입으로 변환
mean_rating_by_user_list_array.shape
```

```python
# 파일에 저장 .savetxt("저장경로", 저장할array, 포맷, delimiter=구분인자)
np.savetxt("./data/2-04) mean_rating_by_user.csv", mean_rating_by_user_list_array, fmt="%.3f", delimiter=',')
```
