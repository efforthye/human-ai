# 년도 별 남자와 여자의 총 합 구하기

```python
total_birth = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
```

```python
total_birth.head(10)
```

```python
total_birth.plot(title='Total births by sex and year')
```
