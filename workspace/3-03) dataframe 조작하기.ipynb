```python
import pandas as pd
import numpy as np
```

```python
## numpy타입으로 6행 4열 랜덤 데이터 만들고 데이터 프레임으로 변환
df = pd.DataFrame(np.random.randn(6,4))
df
```

```python
## 6일간의 데이터를 인덱스 값으로 만들어서 뽑아본다.
df.columns = ['a','b','c','d']
df.index = pd.date_range('20250501', periods=6)
df.index
```

```python
df
```

```python
## f컬럼에 여러가지 값들 추가
df['f']=[1.0, np.nan, 3.5, 6.1, np.nan, 7.0]
df
```

```python
## NAN 값만 제거하기 (how='any'는 nan이 하나라도 포함되어 있으면 그 행을 다 지우는 것이다.)
df.dropna(how='any')
```

```python
## nan이 행 전체에 있을 경우 해당 행을 삭제하기
df.dropna(how='all') 
```

```python
## nan 값을 대체하기 (5로 대체한다.)
df.fillna(value=5.0)
```

```python
## 혹은 불리언 마스크로 값을 제어한다.
df.isnull() # nan 값이 있는것만 true로 표기된다. (nan 값이 있는지 확인)
```

```python
## f열에서 NAN 값이 있는 모든 행을 가져온다.
df.loc[df.isnull()['f'], :]
```

```python
## 인덱스(현재는 날짜로 되어있음) 기준으로 행을 삭제한다.
pd.to_datetime('20250505') #인덱스가 해당날짜인 것을 가져옴
```

```python
## 인덱스가 5월 5일인 행을 삭제한다. (날짜는 데이트타임으로 인덱싱해야한다)
df.drop(pd.to_datetime('20250505')) 
```

```python
# 인덱스 기준으로 두개 이상의 행(5/4, 5/6일)을 삭제한다.
df.drop([pd.to_datetime('20250504'), pd.to_datetime('20250506')])
```

```python
# f열 전체 삭제하기
df.drop('f', axis=1)
```

```python
## 여러 열 한번에 삭제하기
df.drop(['b', 'd'], axis=1)
```

```python
## 데이터 만들기
data = [
    [1.40, np.nan],
    [7.10, -4.5],
    [np.nan, np.nan],
    [0.75, -1.3]
]
data
```

```python
## 데이터 프레임으로 변환하기
df = pd.DataFrame(data, columns = ['one','two'], index=['a','b','c','d'])
df
```

```python
## [중요] 행방향으로 열의 합을 계산한다 ==> nan값은 배제하고 계산된다.
df.sum(axis=0)
```

```python
## 열방향으로 각 행의 합을 계산
df.sum(axis=1)
```

```python
## 특정 열 또는 행의 값 계산 (one 컬럼에 대해 sum)
df.one.sum()
```

```python
df.loc['b'].sum()
```

```python
## two 행에 대해 평균 계산
df.mean(axis=0)['two']
```

```python
## 열에 대해 평균 내는데, nan값이 있으면 그 행은 스킵하고 계산하기
df.mean(axis=1, skipna=False)
```

```python
## one_mean 변수에 'one' 행의 평균값을 넣는다.
one_mean = df.mean(axis=0)['one']
two_min = df.min(axis=0)['two'] # two의 최소값
```

```python
## g
df.one = df['one'].fillna(value=one_mean)
df.two = df['two'].fillna(value=two_min)
df
```
