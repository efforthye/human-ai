pandas에서는 고유하게 정의한 자료 구조인 Series와 DataFrame을 사용하여 빅데이터 분석을 할 수 있다.

# 1) Series

```python
import pandas as pd
import numpy as np
```

# 2) DataFrame
- DataFrame은 pd.DataFrame() 함수를 사용하여 정의한다. (엑셀처럼) DataFrame에 입력할 데이터는 Python 딕셔너리 혹은 numpy의 2차원 array 등의 형태가 될 수 있다.
  

```python
data = {
    "names": ['Hyerim','Hyerim','Hyerim','Charles','Charles'],
    "year": [2014,2015,2016,2015,2016],
    "points": [1.5, 1.7, 3.6, 2.4, 2.9]
}
df = pd.DataFrame(data)
df
```

```python
df.index
```

```python
df.columns
```

```python
df.values
```

```python
# describe 함수
각 컬럼 평균, 분산, 최소/최대값 등 기본 통계량을 산출한 결과를 보여준다.
```

```python
df.describe() # 숫자 있는 컬럼값들 에 각종 통계값 기본적으로 보여준다!!
```

```python
df2 = pd.DataFrame(data, columns = ['year','names','points','penalty'], index = ['1','2','3','4','5'])
df2
```

```python
df2.index
```

```python
df2.columns
```

```python
df2.values
```
